```python
import pandas as pd
import requests
from datetime import datetime, date
import matplotlib.pyplot as plt
```

```python
year = 1998
response_birthYear = requests.get(f'http://127.0.0.1:8000/stars/by_birthyear/{year}')
data_json_birthYear = response_birthYear.json()
dfBirthYear = pd.DataFrame(data_json_birthYear)
dfBirthYear.set_index('id', inplace=True)
dfBirthYear
```

```python
partTitle = 'Adventure'
dfMovieByPartTitle = pd.read_json(f'http://127.0.0.1:8000/movie/by_partTitle?searchTitle={partTitle}')
dfMovieByPartTitle.set_index('id', inplace=True)
dfMovieByPartTitle
```

```python
min_count = 30
response_stat_actor = requests.get(f'http://127.0.0.1:8000/stars/stat_movies?min_count={min_count}')
data_json_stat_actor = response_stat_actor.json()

data_stat_actor = [ {**actor_dict, 'year_first_movie': minYear,'year_flast_movie': maxYear,'movie_count': countMovie }  
         for actor_dict, minYear,maxYear,countMovie in data_json_stat_actor ]
dfStatActor = pd.DataFrame(data_stat_actor)
dfStatActor       
        
```
