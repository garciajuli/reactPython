```python
import pandas as pd
import requests
from datetime import datetime, date
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.dates as mdates
```

```python
year = 1998
response_birthYear = requests.get(f'http://127.0.0.1:8000/stars/by_birthyear/{year}')
data_json_birthYear = response_birthYear.json()
dfBirthYear = pd.DataFrame(data_json_birthYear)
dfBirthYear.set_index('id', inplace=True)
dfBirthYear
```

```python
partTitle = 'Adventure'
dfMovieByPartTitle = pd.read_json(f'http://127.0.0.1:8000/movie/by_partTitle?searchTitle={partTitle}')
dfMovieByPartTitle.set_index('id', inplace=True)
dfMovieByPartTitle
```

```python
min_count = 30
response_stat_actor = requests.get(f'http://127.0.0.1:8000/stars/stat_movies?min_count={min_count}')
data_json_stat_actor = response_stat_actor.json()

data_stat_actor = [ {**actor_dict, 'year_first_movie': minYear,'year_flast_movie': maxYear,'movie_count': countMovie }  
         for actor_dict, minYear,maxYear,countMovie in data_json_stat_actor ]
dfStatActor = pd.DataFrame(data_stat_actor)
dfStatActor       
        
```

```python
dfAllMovies = pd.read_json(f'http://127.0.0.1:8000/movies/count_by_year')
dfAllMovies = dfAllMovies.set_index('year')
dfAllMovies

dfAllStars = pd.read_json(f'http://127.0.0.1:8000/stars?limit=5000000',convert_dates=['birthdate'])
dfAllStars = dfAllStars.set_index('id')

fig, ((ax1, ax2)) = plt.subplots(2, 1, figsize=(16,10))

statStarsByYearToPlot = dfAllStars[~dfAllStars.birthdate.isna()].groupby(pd.Grouper(key='birthdate', freq='Y')).count()
ax1.bar(statStarsByYearToPlot.index,statStarsByYearToPlot.name, color='red', width=250)
ax1.set_xlim(datetime(1845,1,1), datetime(2020,1,1))
ax1.xaxis.set_minor_locator(mdates.YearLocator(base=10))
ax1.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))

ax2.bar(dfAllMovies.index,dfAllMovies.countMovies, color='red')
ax2.set_xlim(1915, 2020)
ax2.xaxis.set_minor_locator(ticker.MultipleLocator(10))
```

```python

min_count = 10
response_count_movie_by_actor = requests.get(f"http://127.0.0.1:8000/stars/stat_movies?min-count={min_count}")
data_count_movie_by_actor_json = response_count_movie_by_actor.json()

response_count_movie_by_actor_flatten = [ { **d['actor'],
                      "movie_count": d['movie_count'],
                      "first_movie_year": d['first_movie_year'], 
                      "last_movie_year": d['last_movie_year']} for d,count,firstYear,lastYear in data_count_movie_by_actor_json]
dfActors = pd.DataFrame(response_count_movie_by_actor_flatten)
dfActors['birthdate'] = pd.to_datetime(dfActors['birthdate']).dt.strftime("%d/%m/%Y")
dfTop10Actors = dfActors[:10] if dfActors.size >= 10 else dfActors
dfTop10Actors
```
